```python
def keys(dic,key):
    if dic.get(key) is None:
        print(f"key you entered was {key}")
        raise KeyError
    return dic.get(key)
try:
    d={"name":"aditya","age":20}
    key=input("enter key")
    print(keys(d,key))

except KeyError as e:
    print(f"{e}enter correct key")

except Exception as e:
    print(f"{e}")

finally:
    print("program executed....")
```

```python
def temp(temp_cel):
    if temp_cel<-273.15:
        raise ValueError
    temp_faren=(temp_cel*9/5)+32
    return temp_faren

try:
        temp_cel=float(input("enter temperatur in celcius"))
        temp_faren=float(temp(temp_cel))
        print(f"temperature in celcius {temp_cel} and farenheit {temp_faren} is")

except ValueError as e:
        print("please enter correct value",e)

except Exception as e:
        print(e)

else:
       print("exception raised")

finally:
        print("program is executed successfully")
```

```python
def temperature(celcius):
    if celcius<-273.15:
        raise ValueError
    if not isinstance(celcius,  (int, float)):
        raise TypeError
    fahrenheight=(celcius*9/5)+32
    return fahrenheight

try:
    celcius=float(input("enter temperature in celcius"))
    print(f"temperature in farenheight is {temperature(celcius)}")

except ValueError as e:
    print(f"{e} enter correct temperature in celcius")

except TypeError as e:
    print(f"{e} enter an integer value")

except Exception as e:
    print(e)

finally:
    print("program executed successufully")
```
