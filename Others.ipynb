```python
#finding area of triangle by using the hero's formula
def heron(a,b,c):
    s=(a+b+c)/2
    area=(s*(s-a)*(s-b)*(s-c))**0.5
    return area
a=float(input("enter side a"))
b=float(input("enter side b"))
c=float(input("enter side c"))
print("sides of triangle are",a,b,c)
print("area of triangle by using heron's formula is",heron(a,b,c))
```

```python
counter = 0  # The counter starts at 0

# Step 2: Define a function to increment the counter conditionally
def conditional_increment(value):
    global counter
    if value >= 1:
        counter = counter + value
        print(f"counter incremented by {value}. new counter variable", counter)
    else:
        print("increment value must be positive")

# Step 3: Call the function with different values to test the behavior
conditional_increment(5)   # Should increment the counter by 5
conditional_increment(-3)  # Should print an error message
conditional_increment(10)  # Should increment the counter by 10

# Printing the final value of the counter after all function calls
print("Final counter value:", counter)
```
