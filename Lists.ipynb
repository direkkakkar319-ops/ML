```python
marks=[67,45,90,100,56]
total=0
count=0

for i in range(5):
    if marks[i]>50:
        count=count+1
    total=marks[i]+total
greatest=marks[0]
lowest=marks[4]

for i in marks:
    if i>greatest:
        greatest=i
    elif i<lowest:
        lowest=i

print("total marks",total) 
print("students abovew 50 marks",count)
print("greatest number",greatest)
print("lowest number",lowest)
```
