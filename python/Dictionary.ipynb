```python
se = "apple banana apple banana orange mango apple"
se = se.strip()
words = se.split()     # split into list of words
fre = {}               # frequency dictionary
for w in words:
    fre[w] = fre.get(w, 0) + 1
print(fre)
```

```python
#removing and deleting elements in dictionary

d1={1:"ejfer", 2:"egege", 3:535,"age":"hello"}
print(d1)

#1 popping
d1.pop(2)
print(d1)  

#2 deleting element from dictionary
del d1[1]
print(d1)

#3 popitem
lol ,luli=d1.popitem()
print("key is {},value is {}".format(lol,luli))
print(d1)

#4 clearing the dictionary
d1.clear()
print(d1)
```

```python
def gg(words):
    result = {}
    for word in words:
        first_letter = word[0].lower()
        if first_letter not in result:
            result[first_letter]=set()
        result[first_letter].add(word)
    return result
arr=["cherry","chole","blueberry","banana","apricot","apple"]
print(gg(arr))
```

```python
#PRINTING DICTIONARY IN DIFFRENT STYLES
d1={1:"a",2:"a",3:"a",4:"a",5:"a",6:"a",7:"a",8:"a",9:"a"}

#keys only
for key in d1:
    print(key)

#values only
for value in d1.values():
    print(value)

#keys and values
for key, value in d1.items():
    print(key,value)
```
