```python
def vowels(k):
    count = 0
    print(k)
    for i in k:
        if i in ['a', 'e', 'i', 'o', 'u']:
            count += 1
    return f"number of vowels in the string are {count}"

st = input("Enter a string: ").lower()
print(vowels(st))
```

```python
def length(st):
    st_new=st.strip().split()
    result={}
    for word in st_new:
        result[word]=len(word)
    for word,size in result.items():
        print( f"{word}:{size}")
string=str(input("enter a string"))
length(string)
```
