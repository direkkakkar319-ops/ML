```python
# ===== Example 7: Employee with bonus calculation =====
class Employee:
    def __init__(self, name, salary, department):
        self.name = name
        self.salary = salary
        self.department = department

    def bonus(self):
        # Bonus based on department
        if self.department == "Sales":
            self.salary = self.salary * 1.1
        else:
            self.salary = self.salary * 1.05

    def display(self):
        print(f"Updated salary is {self.salary}")

employee1 = Employee("Rohan", 10000, "Sales")
employee1.bonus()
employee1.display()
```

```python
# ===== Example 12: Temperature conversion using static methods =====
class Temperature:

    @staticmethod
    def to_fahrenheit(temp):
        return 1.8 * temp + 32

    @staticmethod
    def to_celsius(temp):
        return (temp - 32) / 1.8

print(Temperature.to_fahrenheit(3))
print(Temperature.to_celsius(10))
```

```python
# ===== Example 13: Static and class methods in Car class =====
class Car:
    total_car = 0

    def __init__(self, model, mileage):
        self.model = model
        self.mileage = mileage

    @staticmethod
    def is_mileage(obj):
        if obj.mileage > 15:
            return "\nmilage is good"
        else:
            return "\nmilage is bad"

    @classmethod
    def show_total(cls, total_car):
        print(f"Total cars are {total_car}")

c1 = Car("X8-oiur", 29)
print(Car.is_mileage(c1))

c1.show_total(20)

c2 = Car("BMW-xB", 10)
print(Car.is_mileage(c2))

c2.show_total(1)
```

```python
# ===== Example 19: Using super() in multi-level inheritance =====
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

class Ebook(Book):
    # here we use super as we have to use the varibles of constructor function but not any function
    def __init__(self, title, author, file_size):
        super().__init__(title, author)  # called as parent constructor
        self.file_size = file_size

    def display(self):
        print(self.title)
        print(self.author)
        print(self.file_size)

ebook = Ebook("Python", "Author1", 23)
ebook.display()
```

```python
# Import ABC and abstractmethod to create abstract base classes
from abc import ABC, abstractmethod

# ===== Example 6: Shapes and area calculation =====
class Shape(ABC):

    # Abstract method: must be implemented by subclasses
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    # Implement abstract method
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Implement abstract method
    def area(self):
        return self.length * self.width

# Create objects and calculate area
c = Circle(5)
r = Rectangle(3, 4)
print("Circle area:", c.area())
print("Rectangle area:", r.area())
```
